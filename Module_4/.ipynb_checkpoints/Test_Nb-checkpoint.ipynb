```python
def handle_missing_values(df):
    for i in df.columns:
    #print(df[i].dtype)
        if df[i].dtype in ['int32','float64','int64']:
            df[i].fillna(df[i].mean(), inplace=True)
        elif df[i].dtype == 'object':
            df[i].fillna(df[i].mode().name, inplace=True)
            #print(f"{i}:{df[i].isna().sum()}")
return df
```

```python
import pandas as pd
import numpy as np
try:
    # Read the traffic crashes CSV file and store it in a dataframe
    df_crashes = pd.read_csv("data/traffic_crashes.csv")
    df_crashes = handle_missing_values(df_crashes)
    # Read the traffic crash vehicle CSV file and store it in a dataframe
    df_vehicles = pd.read_csv("data/traffic_crash_vehicle.csv")
    df_vehicles = handle_missing_values(df_vehicles)
    # Read the traffic crash People CSV file and store it in a dataframe
    df_people= pd.read_csv("data/traffic_crash_people.csv")
    df_people = handle_missing_values(df_people)
except FileNotFoundError as e:
    # Handle exception if any of the files are missing
    print(f"Error: {e}")
except Exception as e:
    # Handle any other exceptions
    print(f"Error: {e}")
    
```

```python
def transform_data(df,df2,df3):
    # drop duplicate rows
    df = df.drop_duplicates()

    print(f"df_crash:{df.shape}")
    print(f"df2_vehicles:{df2.shape}")
    print(f"df3_people:{df3.shape}")
    # convert columns to appropriate data types
    try:
        df['crash_date'] = pd.to_datetime(df['crash_date'], format='%m/%d/%Y')
    except:
        pass

    try:
        df['posted_speed_limit'] = df['posted_speed_limit'].astype('int32')
    except:
        pass

    # merge the three dataframes into a single dataframe
    merge_01_df = pd.merge(df, df2, on='crash_record_id')

    # Merge the DataFrames
    merge_01_df = pd.merge(df, df2, how='inner', on='crash_record_id', suffixes=('', '_remove'))
 
    # remove the duplicate columns
    merge_01_df.drop([i for i in merge_01_df.columns if 'remove' in i],
               axis=1, inplace=True)

    all_data_df = pd.merge(merge_01_df, df3,how='inner', on='crash_record_id', suffixes=('', '_remove'))
    
    # remove the duplicate columns
    all_data_df.drop([i for i in all_data_df.columns if 'remove' in i],
               axis=1, inplace=True)
    
    print(f"all_data_df:{all_data_df.shape}")
    return all_data_df
    
```

```python
df = transform_data(df_crashes,df_vehicles,df_people)
```
